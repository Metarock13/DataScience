# Предобработка данных

```python
#приводим названия колонок к нижнему регистру
games.columns = games.columns.str.lower()

#просмотр пропусков по столбцам
games.isna().sum()
games.head()
```

```python
print(len(games))
games[games['name'] == 'Need for Speed: Most Wanted']
```

```python
#проверка на дубликаты
games.duplicated().sum()
```

```python
print(len(games))
games[games['name'] == 'Need for Speed: Most Wanted']
```

```python
#удаляем строку без названия игры, так как там почти все данные пустые
games.drop(labels = [659],axis = 0,inplace=True)
```

```python
# удаляем строки без даты (важнейшей метрики проекта)
games.dropna(axis=0,inplace=True,subset=['year_of_release'])

# добавляем столбец с суммарными продажами для каждой игры
games['total_sales'] = games.na_sales + games.eu_sales + games.jp_sales + games.other_sales
```

```python
#меняем десятичный тип на целочисленный, так как год и оценка критиков здесь не подразумевают доли + меняем строковый тип на десятичный для рейтинга пользователей, так как нам нужны числа, а не текст
columns_to_retype = ['year_of_release','critic_score','user_score']
for vid in columns_to_retype:
    games[vid]=pd.to_numeric(games[vid], downcast='integer',errors='coerce')

print(games['user_score'].corr(games['critic_score']))
```

```python
games.loc[games['rating']=='K-A', 'rating']='E'  # К-А и Е - это одни и те же возрастные рейтинги
games['rating']=games['rating'].fillna(value='unknown') # заполняем пропуски в rating на unknown

games.isna().sum()
```

Был добавлен столбец total_sales с суммарными продажами для каждой игры.

Пропуски в столбцы возрастного рейтинга заполним пометкой unknown. К тому же, так как К-А и Е - это одни и те же возрастные рейтинги, можно проставить вместо первого показателя Е.

В столбце user_score обнаружено tbd - аббревиатура от английского To Be Determined (будет определено). Значение превращается в пропуск после приведения столбца к числовому виду. Чтобы заполнить пропуски в указанных столбцах, можно воспользоваться категоризацией и медианными значениями, так как прослеживается высокий показатель корреляции между столбцами (0,58)

Причины, которые могли привести к пропускам в данных: технические ошибки, человеческий фактор, сложности в расчете. Например, не удалось получить точные данный по оценочному рейтингу игры, так как она очень старая или непопулярная. Возрастной рейтинг у разных игр на различных платформах в разных странах может отличаться по тем или иным причинам.

<font color='Blue'><b>2. Выявим за какой характерный срок появляются новые и исчезают старые платформы?</b></font>

```python
#обзор продаж по платформам
games_plats=games.pivot_table(index='platform', values='total_sales',aggfunc=['sum'])
games_plats.columns=['sales']
top_plats=games_plats.sort_values(by='sales',ascending=False).head(10)
top_list=list(top_plats.index[:10])
top_data=games.loc[games['platform'].isin(top_list)]
games_stats=top_data.pivot_table(index='platform', columns='year_of_release',values='total_sales',aggfunc=['sum'])
games_stats_t = games_stats.T #упорядочивание столбцов

#распределение по годам
games_stats_t.plot(xlabel='year_of_release',ylabel='total_sales' ,grid=True,figsize=(8, 5),title='Рейтинг продаж топ-10 платформ')
```

Можно заметить, что новые платформы появляются раз в два года, а если речь идет об одной компании, то каждые шесть лет. С другой стороны, старые платформы пропадают через 9-10 лет.

Изучение фокусируется на периоде между 2013 и 2016 годами, так как именно с 2013 года на рынке появилась острая конкуренция, сопровождающаяся уходом старых платформ.

```python
#актуальизируем данные
games_stats_t=games_stats_t.query('year_of_release>=2013')

#отслеживание динамики продаж
games_stats_t.plot(xlabel='year_of_release', ylabel='total_sales',grid=True,fontsize=10,figsize=(10, 8),title='Рейтинг продаж лучших платформ c 2013 года')
games_stats_t
```

# Проверка гипотез
H_0: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

H_a: Средние пользовательские рейтинги платформ Xbox One и PC различаются

```python
alpha = .05

results = st.ttest_ind(
    games2013[games2013['platform'] == 'XOne']['user_score'],
    games2013[games2013['platform'] == 'PC']['user_score'],
    equal_var=False,
    nan_policy='omit')

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Шансы случайно получить различие такого или большего уровня составляют почти 15%, что говорит о том, что нельзя сделать вывод о значимом различии между рейтингами.

**Гипотеза подтверждена: средние пользовательские рейтинги платформ Xbox One и PC схожи**

Для удобства сделаем аналогичную проверку по жанрам Action и Sports, но вывод сделаем относительно альтернативного исхода.

H_0: Средние пользовательские рейтинги жанров Action и Sports одинаковые

H_а: Средние пользовательские рейтинги жанров Action и Sports разные

```python
alpha = .05

results = st.ttest_ind(
    games2013[games2013['genre'] == 'Action']['user_score'],
    games2013[games2013['genre'] == 'Sports']['user_score'],
    equal_var=False,
    nan_policy='omit')

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Вероятность случайно получить такое или большее различие крайне мало.

**Гипотеза подтверждена: рейтинги жанров Action и Sports разные**

Критерий для проверки гипотез сводится к сравнению pvalue и уровня критической значимости, так как нулевая гипотеза отвергается, если вероятность получить то среднее, которое имеется в выборочных данных, слишком мала.
