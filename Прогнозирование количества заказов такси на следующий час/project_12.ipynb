#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

```python
df = pd.read_csv('datasets/taxi.csv', index_col = [0], parse_dates = [0])
df.sort_index(inplace=True)
df.head()
```

Визуализируем полученные результаты на графике

Из-за большого объема данных в датасете, нормальная визуализация графика сезонности количества заказов становится затруднительной, так как может привести к заполнению изображения большим количеством линий.
```python
df_resampled_week = df_resampled['2018-03-10 01:00:00':'2018-03-17 01:00:00']
decomposed_week = seasonal_decompose(df_resampled_week)
```

```python
decomposed_week.seasonal.plot(figsize=(12,8))
plt.show()
```

В течение дня количество заказов постепенно растет, достигая пика к вечеру, когда люди становятся более активными - например, возвращаются с работы или устраивают встречи с друзьями и родственниками.

Создадим график "шума", который представляет собой компонент данных, не связанный ни с сезонностью, ни с трендом

Здесь мы видим разнообразные факторы, которые могут влиять на количество заказов такси. Это может включать в себя такие события, как внезапный ливень, который подразумевает возможность добраться до дома сухими в автомобиле, а также сильные пробки, которые заставляют людей пересаживаться на общественный транспорт, например, метро.

## Обучение

###  Формирование обучающих признаков

Разработаем функцию для создания признаков на основе значений месяца, числа и дня недели (не включая год, так как все данные относятся к 2018 году).

```python
def make_features(data, max_lag, rolling_mean_size):
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour

    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
```

```python
make_features(df_resampled, 100, 10)
df_resampled.head()
```

```python
from statsmodels.graphics import tsaplots

taxi = pd.read_csv("datasets/taxi.csv", index_col=[0], parse_dates=[0]).resample('1H').sum()
ts = taxi['num_orders']

fig = tsaplots.plot_acf(ts,lags= 190)
fig.set_size_inches(18.5, 6)
plt.ylim(-.1, 1.1)
plt.show() 
```

### Разбиение на обучающую и тестовую выборки, удаление пропусков

Разделим датасет на тренировочную и тестовую выборки в соотношении 9:1, с учетом сохранения порядка ряда. Для этого установим параметр shuffle равным False.

```python
df_train, df_test = train_test_split(df_resampled, test_size = 0.1, shuffle = False, random_state = 12345)
```

После добавления временного лага в признаки мы обнаружили пропуски в данных, и решили удалить их.

```python
df_train.dropna(inplace=True)

df_train.head()
```

```python
df_test.head()
```

Выполняем проверку правильности разбиения данных на выборки

```python
print("train:", df_train.index.min(), df_train.index.max())

print("test:", df_test.index.min(), df_test.index.max())
```

### Отделение обучающих и целевого признаков от датасета

```python
features_train = df_train.drop(['num_orders'], axis=1)
target_train = df_train['num_orders']

features_test = df_test.drop(['num_orders'], axis=1)
target_test = df_test['num_orders']
```

### Обучение и валидация модели DecisionTree

```python
features_train.shape
```

```python
features_test.shape
```

```python
features_train.shape[0] // (8+1)
```

Для проведения кросс-валидации с использованием метода гридсерч будем использовать параметр n_splits, который можно определить, используя формулу, указанную в документации test_size = n_samples // (n_splits + 1) --> 442 = 3924 // (n_splits + 1) --> n_splits ~ 8

Начнем с использования простого обучающего дерева. Чтобы найти оптимальные параметры, мы применим метод GridSearch, в котором будет использоваться специальная кросс-валидация, адаптированная для временных рядов.

```python
%%time
model_tree = DecisionTreeRegressor()

param_grid = { 'max_depth': [1,30], 'min_samples_leaf': [1, 30],}

tscv = TimeSeriesSplit(n_splits=8)

gs = GridSearchCV(model_tree, param_grid=param_grid, scoring='neg_mean_squared_error', n_jobs=-1, cv=tscv, verbose=1)

model_tree_fitted = gs.fit(features_train, target_train)
```

Выведем лучший результат метрики и подобранные гиперпараметры модели, которые были получены в результате нашего анализа

```python
rmse_tree_train = (model_tree_fitted.best_score_*(-1))**0.5
print(rmse_tree_train)
print(model_tree_fitted.best_params_)
```

### Обучение модели LinearRegression

Создаем модель линейной регрессии, обучаем ее на тренировочных признаках и получаем предсказания на основе этой модели

```python
%%time
model_LR = LinearRegression()

param_grid = {'fit_intercept':[True, False],'copy_X':[True, False]}

tscv = TimeSeriesSplit(n_splits=8)
gsl = GridSearchCV(model_LR, param_grid=param_grid, scoring='neg_mean_squared_error', n_jobs=-1, cv=tscv, verbose=1)

model_LR_fitted = gsl.fit(features_train, target_train)
rmse_LR_train = (model_LR_fitted.best_score_*(-1))**0.5
print(rmse_LR_train)
print(model_LR_fitted.best_params_)
```

### Подбор гиперпараметров модели RandomForest

Проведем подбор оптимальных параметров для обучения модели случайного леса

```python
%%time
model_forest = RandomForestRegressor()

param_grid = { 'max_depth': [1,20], 'n_estimators': [1,100], }

tscv = TimeSeriesSplit(n_splits=8)
gs = GridSearchCV(model_forest, param_grid=param_grid, scoring='neg_mean_squared_error', n_jobs=-1, cv=tscv, verbose=1)

model_forest_fitted = gs.fit(features_train, target_train)
```

Выведем лучший результат метрики и подобранные гиперпараметры модели, которые были получены в результате нашего анализа

```python
rmse_forest_train = (model_forest_fitted.best_score_*(-1))**0.5
print(rmse_forest_train)
print(model_forest_fitted.best_params_)
```

### Подбор гиперпараметров модели LightGBM

Далее добавим модель градиентного бустинга в нашу работу.

```python
%%time
train_dataset = lgb.Dataset(features_train, label=target_train)
test_dataset = lgb.Dataset(features_test, label=target_test)

parameters = {'max_depth': np.arange(10, 15), 'n_estimators': np.arange(45, 50) }

tscv = TimeSeriesSplit(n_splits=8)
gs = GridSearchCV(lgb.LGBMRegressor(random_state=12345), param_grid=param_grid, scoring='neg_mean_squared_error', n_jobs=-1, cv=tscv, verbose=1)

gs.fit(features_train, target_train)

gs.best_params_
```

Рассмотрим значение метрики RMSE для модели с выбранными гиперпараметрами

```python
rmse_lightgbm_train = ((gs.best_score_)*-1)**0.5
print('RMSE LightGBM: {:.2f}'.format(rmse_lightgbm_train))
```

### Подбор гиперпараметров CatboostRegressor

Создадим модель CatBoost и обучим ее на наших данных

```python
%%time
cbr = CatBoostRegressor(iterations=300, random_state=12345, loss_function='RMSE')

cbr.fit(features_train, target_train,  verbose=30)
preds_train = cbr.predict(features_train)
```

Вычислим значение RMSE для данной модели.

```python
mse_catboost_train = mean_squared_error(target_train, preds_train)
rmse_catboost_train = mse_catboost_train ** 0.5
rmse_catboost_train
```

## Тестирование

```python
%%time
tree = DecisionTreeRegressor(max_depth = model_tree_fitted.best_params_['max_depth'], min_samples_leaf=model_tree_fitted.best_params_['min_samples_leaf'], random_state=12345)

tree.fit(features_train, target_train)
```
